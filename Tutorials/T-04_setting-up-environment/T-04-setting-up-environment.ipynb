## If working locally on your Deskstop, you can use your terminal to change directories and set the right environments via conda.  
## These prompts will allow you to manage and resolve any issues:

- conda —version # will tell you base version of conda

- conda env create —file environment.yml # creates environment in conda and base modules/packages

- conda env list # will list environments within conda module

- conda activate playenv # will activate said environment that is listed and will change that from your base environment to playenv

- import # import will bring a module into a environment, for example import pandas

- conda deactivate # will deactivate current environment and return to base environment
